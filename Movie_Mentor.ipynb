```python
def castCon(list1):
    c=0
    L=[]
    for i in list1:
        if c!=4:
            L.append(i['name'])
            c=c+1
        else:
            break
            
    return L
        
```

```python
def convertCrew(list2):
    L=[]
    for i in list2:
        if i['job']=='Director':
            L.append(i['name'])
            break
            
    return L
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
stop_words=set(stopwords.words('english'))
```

```python
def removingSWandDual(obj):
    word_tokens=word_tokenize(obj)
    filter_sen=[w for w in word_tokens if not w.lower() in stop_words]
    remov_d_text=[ps.stem(w) for w in filter_sen]
    return ' '.join(remov_d_text)
```

```python
new_df.loc[:,'tags']=new_df['tags'].apply(removingSWandDual)
```
