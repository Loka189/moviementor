```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
movies.info()
```

```python
movies=movies[['movie_id','title','genres','overview','release_date','budget','revenue','runtime','cast','crew']]
```

```python
movies.info()
```

```python
import ast
movies['genres']=movies['genres'].apply(lambda x:ast.literal_eval(x))
movies['cast']=movies['cast'].apply(lambda x:ast.literal_eval(x))
movies['crew']=movies['crew'].apply(lambda x:ast.literal_eval(x))
```

```python
def convert(List):
    L=[]
    for i in List:
        L.append(i['name'])
        
    return L
```

```python
movies.loc[:,'genres']=movies['genres'].apply(convert)
```

```python
def castCon(list1):
    c=0
    L=[]
    for i in list1:
        if c!=4:
            L.append(i['name'])
            c=c+1
        else:
            break
            
    return L
```

```python
movies.loc[:,'cast']=movies['cast'].apply(castCon)
```
